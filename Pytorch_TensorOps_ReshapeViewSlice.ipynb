```python
import torch
import numpy as np
mytorch = torch.arange(10)
mytorch
## RESHAPE & VIEW Tensor Operations
mytorch = mytorch.reshape(2,5)
mytorch
```

```python
## Reshape if you don't know the number of items using "-1"
mytorch2 = torch.arange (15)
mytorch2 = mytorch2.reshape(3,-1)
mytorch2
mytorch3 = mytorch2.view (3,5)
mytorch3
```

```python
## With reshape and view they will update if the underlying Tensor changes
mytorch5 = torch.arange (10)
mytorch5[1] = 31415
mytorch6 = mytorch5.reshape(2,5)
mytorch6
```

```python
## SLICES
mytorch7 = torch.arange (10)
mytorch7[7]
## GRAB A SLICE
mytorch8 = mytorch7.reshape (5,2)
mytorch8[:,1]
## RETURN A COLUMN
mytorch8[:,1:]
```
