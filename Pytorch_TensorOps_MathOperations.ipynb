```python
import torch
import numpy as np
## Tensor Math Operations: Add, Subtract, Multiply, Divide, Remainders, Exponents
## Shorthand, Longhand, Reassignment
tensor_1 = torch.tensor([1,2,3,4])
tensor_2 = torch.tensor([5,6,7,8])
## Addition
tensor_1 + tensor_2
```

```python
## Sub Fuction (Subtraction)
torch.sub(tensor_2,tensor_1)
```

```python
## Addition Longhand
torch.add (tensor_1, tensor_2)

## subtraction
tensor_2-tensor_1
```

```python
## Multiplication
tensor_2 * tensor_1
## Multiplication Longhand
torch.mul (tensor_2, tensor_1)
```

```python
## Division
torch.div(tensor_2, tensor_1)
tensor_2/tensor_1
```

```python
## Remainder Modulus
tensor_2 % tensor_1
torch.remainder (tensor_2, tensor_1)
```

```python
## Exponents / Power
torch.pow(tensor_1, tensor_2)
```

```python
## Another way to write longhand
tensor_1.add (tensor_2)
```

```python
## Reassignment
tensor_1 + tensor_2
tensor_1
## Tensor_1 = Tensor_1 + Tensor_2
tensor_1.add_(tensor_2)
tensor_1
```
